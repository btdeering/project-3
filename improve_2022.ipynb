```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)

# confirm that our new database was created
print(mongo.list_database_names())
```

```python
# assign the covid database to a variable name
db = mongo['covid_2022']
```

```python
# assign the collection to a variable
nations = db['nations']
```

```python
query = {'Country_Region': 'US'}
results = nations.find(query)
sort = [("Province_State", 1)]
limit = 5
results = list(nations.find(query).sort(sort).limit(limit))

pprint(results)
```

```python
output = []
results = nations.find(query)
for result in results:
    output.append(result)

usa_df = pd.DataFrame(output)
# Display the number of rows in the DataFrame
usa_df.drop(columns= ['Recovered','Active'], inplace=True)# Display the first 10 rows of the DataFrame
usa_df.dropna(axis=0, inplace=True)
print("Rows in DataFrame: ", len(usa_df))
usa_df.head()
```

```python

states_group = usa_df.groupby(['Province_State'])
states=states_group.mean()
states_df = states[['Lat','Long_','Case_Fatality_Ratio']]
states_df
```

```python
states_group = usa_df.groupby(['Province_State'])
states = states_group.agg({'Region': 'first', 'Lat': 'mean', 'Long_': 'mean','Confirmed': 'sum', 'Deaths':'sum', 'Case_Fatality_Ratio': 'mean'})
states.rename(columns={'Long_': 'long'}, inplace=True)
states.rename(columns={'Lat	': 'lat'}, inplace=True)
states.rename(columns={'Confirmed': 'total_confirmed'}, inplace=True)
states.rename(columns={'Deaths': 'total_deaths'}, inplace=True)
states.rename(columns={'Case_Fatality_Ratio': 'average_case_fatality_ratio'}, inplace=True)

states_df = states.reset_index()
states_df.rename(columns={'Province_State': 'State'}, inplace=True)
states_df = states_df.round(3)

states_df
```

```python
# Assuming result_df is a pandas DataFrame
results = states_df.to_dict(orient='records')

# Convert results to JSON string
results_json = json.dumps(results, indent=4)

# Define the output file path
output_file_path = 'Improved/improved_2022.json'

# Write the JSON string to the output file
with open(output_file_path, 'w') as output_file:
    output_file.write(results_json)
```
